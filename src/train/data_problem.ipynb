### Modifikasi Data Agar Memenuhi Kriteria

import modul

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

load dataset

```python
path_sample = "../dataset/original_sample.csv"
df = pd.read_csv(path_sample)

df.head()
```

Missing Value
- Hapus secara acak beberapa nilai pada kolom tertentu (misal Units_Sold atau Avg_Price_EUR)

```python
df.loc[df.sample(frac=0.002).index, 'Region'] = np.nan
```

```python
df.loc[df.sample(frac=0.002).index, 'Units_Sold'] = np.nan
```

cek missing value

```python
df.isnull().sum()
```

cek outlier
- Ubah beberapa nilai menjadi sangat ekstrem (misal 10 kali lipat dari nilai normal)

```python
df.loc[df.sample(frac=0.001).index, 'Revenue_EUR'] = df['Revenue_EUR'].max() * 10
```

Cek Tipe tidak konsisten
- Ubah beberapa baris pada kolom Month menjadi teks (misal "Jan"). mengganti sebagian nilai yang memang bernilai 1 menjadi 'Jan', maka sampling harus dilakukan hanya pada baris yang Month == 1. Langkahnya: Ambil baris yang Month == 1, Dari baris itu ambil sebagian secara acak dan Ganti menjadi 'Jan'

```python
idx = df[df['Month'] == 1].sample(frac=0.3).index
df.loc[idx, 'Month'] = 'Jan'

print(idx)
```

Cek Duplikasi
- Gandakan beberapa baris dan tambahkan ke dataframe

```python
# Ambil sample acak 10 baris data, random state biar data nya tidak berubah ubah yang di pick tiap re run.
duplicates = df.sample(n=20, random_state=42)
df = pd.concat([df, duplicates], ignore_index=True)

print(duplicates)
```
