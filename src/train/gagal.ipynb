Kolom Month saat ini masih bertipe string ('1' sampai '12'). Kita perlu mengubahnya menjadi integer dan membuat kolom tanggal untuk memudahkan plotting (visualisasi) dan agregasi(penggabungan data/merangkum).

```python
# Ubah Month menjadi integer
df['Month'] = df['Month'].astype(int)

# Buat kolom Date (awal bulan)
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-01')

# Urutkan berdasarkan tanggal
df = df.sort_values('Date').reset_index(drop=True)

df.head()
```

Penjelasan:
- astype(int) mengubah string angka menjadi integer.
- pd.to_datetime membuat objek datetime yang bisa diurutkan dan di-resample.
- Pengurutan penting untuk analisis deret waktu.

##### **8. Agregasi / Menggabungkan Data Menjadi Satu Nilai**
Menjumlahkan Units_Sold untuk setiap bulan (semua region dan model) untuk mendapatkan total penjualan global per bulan.

```python
monthly_sales = df.groupby(['Year', 'Month'])[['Units_Sold', 'Revenue_EUR']].sum().reset_index()
monthly_sales['Date'] = pd.to_datetime(monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str) + '-01')
monthly_sales = monthly_sales.sort_values('Date').reset_index(drop=True)

print(monthly_sales.head())
```

Data per baris (per model/region) digabungkan menjadi satu baris per bulan.

##### **9. Visualisasi Deret Waktu Bulanan**

```python
import plotly.express as px
import plotly.graph_objects as go

# 9.1 Visualisasi Keseluruhan Tahun
fig = px.line(monthly_sales, x='Date', y='Units_Sold', markers=True,
              hover_data={'Date': '|%B %Y', 'Revenue_EUR': ':,.0f'},
              title='Total Penjualan Bulanan BMW Global (Keseluruhan Tahun 2018-2025)',
              labels={'Units_Sold': 'Unit Terjual', 'Date': 'Tanggal (Bulan/Tahun)'})
fig.update_traces(hovertemplate='<b>Bulan/Tahun</b>: %{x|%B %Y}<br><b>Unit Terjual</b>: %{y:,.0f}<br><b>Pendapatan (EUR)</b>: €%{customdata[0]:,.0f}')
fig.show()
```

```python
# 9.2 Visualisasi Pola Pembelian Per Tahun (2018 - 2025)
fig_yearly = px.line(monthly_sales, x='Month', y='Units_Sold', color='Year', markers=True,
                     hover_data={'Revenue_EUR': ':,.0f'},
                     title='Pola Penjualan Per Tahun (2018 - 2025)',
                     labels={'Month': 'Bulan (1-12)', 'Units_Sold': 'Unit Terjual'})
fig_yearly.update_traces(hovertemplate='<b>Bulan</b>: %{x}<br><b>Unit Terjual</b>: %{y:,.0f}<br><b>Pendapatan (EUR)</b>: €%{customdata[0]:,.0f}')
fig_yearly.update_layout(xaxis=dict(tickmode='linear', tick0=1, dtick=1))
fig_yearly.show()
```

```python
# 9.3 Visualisasi Rata-rata Penjualan Per Bulan (Kenaikan/Penurunan)
monthly_avg = df.groupby('Month')[['Units_Sold', 'Revenue_EUR']].mean().reset_index()
monthly_avg.columns = ['Bulan', 'Rata2_Penjualan', 'Rata2_Pendapatan']

# Cari indeks dengan rata-rata penjualan tertinggi dan terendah
max_idx = monthly_avg['Rata2_Penjualan'].idxmax()
min_idx = monthly_avg['Rata2_Penjualan'].idxmin()

fig_avg = go.Figure()
# Line chart untuk rata-rata penjualan
fig_avg.add_trace(go.Scatter(x=monthly_avg['Bulan'], y=monthly_avg['Rata2_Penjualan'], 
                             mode='lines+markers', name='Rata-rata Penjualan',
                             customdata=monthly_avg['Rata2_Pendapatan'],
                             hovertemplate='<b>Bulan</b>: %{x}<br><b>Rata-rata Terjual</b>: %{y:,.0f}<br><b>Rata-rata Pendapatan</b>: €%{customdata:,.0f}'))

# Highlight titik tertinggi
fig_avg.add_trace(go.Scatter(x=[monthly_avg['Bulan'][max_idx]], y=[monthly_avg['Rata2_Penjualan'][max_idx]],
                             mode='markers', marker=dict(color='green', size=12, symbol='star'),
                             name=f'Tertinggi (Bulan {monthly_avg["Bulan"][max_idx]})'))

# Highlight titik terendah
fig_avg.add_trace(go.Scatter(x=[monthly_avg['Bulan'][min_idx]], y=[monthly_avg['Rata2_Penjualan'][min_idx]],
                             mode='markers', marker=dict(color='red', size=12, symbol='x'),
                             name=f'Terendah (Bulan {monthly_avg["Bulan"][min_idx]})'))

fig_avg.update_layout(title='Rata-rata Penjualan per Bulan (Pola Kenaikan/Penurunan)',
                      xaxis=dict(tickmode='linear', tick0=1, dtick=1, title='Bulan'),
                      yaxis=dict(title='Rata-rata Units Sold'))
fig_avg.show()

print(monthly_avg)
```
